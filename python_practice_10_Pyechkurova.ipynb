# IІ. Data Visualization.

2. (10б) Створіть pie chart, який би показував співвідношення виживших до загиблих.

```python
explode = (0.1, 0)  # Slightly increased explode for the 'Died' section for emphasis
survived = titanic_df['Survived'].value_counts()
fig1, ax1 = plt.subplots()
ax1.pie(survived, labels=['Died', 'Survived'], autopct='%1.1f%%', explode=explode,
        shadow=False, startangle=100)  # Removed shadow, changed start angle
plt.title('Passenger Survival on the Titanic')
ax1.axis('equal')  # Ensure pie is drawn as a circle.
plt.show()
```

*прокоментуйте отримані візуалізації

Ця кругова діаграма показує розподіл пасажирів "Титаніку" за критерієм виживання чи загибелі після аварії. З графіка можна зрозуміти, що значна частина пасажирів, приблизно 62%, не вижила, в той час як 38% вціліли.

3. (20б) Створіть bar chart, який би показував співвідношення загиблих до виживших для кожної статі. Для groupby використовуйте атрибут as_index=False.


image.png

```python
gender_survival_count = titanic_df.groupby(['Sex', 'Survived']).size().reset_index(name='count')
gender_survival_pivot = gender_survival_count.pivot_table(index='Sex', columns='Survived', values='count').fillna(0)
gender_survival_pivot.plot(kind='bar', title='Survival Comparison by Gender', figsize=(7, 5))
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)
plt.legend(title='Survival Status')
plt.show()
```

*прокоментуйте отримані візуалізації

Гістограма ілюструє розподіл виживших після аварії на "Титаніку" за гендерною ознакою. З графіка можна висновувати, що кількість виживших серед жінок значно переважала кількість виживших чоловіків.

5. (10б) Побудуйте наступний графік, дайте йому назву та опишіть, що власне на них відображається.

image.png

```python
sns.violinplot(data=titanic_df, x='Pclass', y='Age', hue='Survived', split=True, palette="muted")
plt.title("Distribution of Passenger Ages by Class and Survival", fontsize=14)
plt.xlabel("Passenger Class", fontsize=12)
plt.ylabel("Age", fontsize=12)
plt.legend(title="Survived", loc='upper right', fontsize='small')
plt.tight_layout()
plt.show()
```

*прокоментуйте отримані візуалізації

Цей графік використовує формат скрипкової діаграми для демонстрації розподілу вікових груп серед пасажирів "Титаніка", з урахуванням їхнього класу (Pclass) та результатів виживання(Survived). Вона дає уявлення про те, як вік, клас та статус виживання пов'язані між собою.

Висновки по графіку:
Загальний розподіл віку: Більшість пасажирів були у віці від 20 до 40 років.

Вплив класу: Пасажири 1-го класу мали значно більшу частку людей похилого віку, ніж 2-го та 3-го класів.

Вплив виживання:
Пасажири 1-го класу мали значно вищу ймовірність виживання, ніж 2-го та 3-го класів, особливо серед чоловіків. Діти та молоді люди мали значно більші шанси на виживання, ніж люди похилого віку

6. (13б) Побудуйте графік розподілу частот по віку пасажирів.


image.png

```python
mean_age = titanic_df['Age'].mean()
median_age = titanic_df['Age'].median()
plt.hist(titanic_df['Age'], bins=20, edgecolor='black', alpha=0.7, color='purple')
plt.axvline(x=median_age, color='purple', linestyle='solid', label='Median')
plt.axvline(x=mean_age, color='purple', linestyle='dashed', label='Mean')
plt.title("Passengers' Ages Distribution", fontsize=14)
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()
plt.show()
```

*прокоментуйте отримані візуалізації

Гістограма показує, що вік пасажирів на Титаніку не має нормального розподілу. Більше пасажирів було у 20-ти та 30-ти роках, ніж у будь-якій іншій віковій групі. Медіанний вік становить 28 років, а середній - 30 років. Це означає, що на Титаніку було більше молодих пасажирів, ніж старших.

Отримані візуалізації надають змогу оглянути взаємозв'язки між різними змінними та виживанням пасажирів.

Висновки, які можна зробити на основі візуалізації:

Клас: Найбільші шанси вижити були у пасажирів першого класу, у другого та третього - менще

Стать: Чоловіки мали менші шанси на виживання ніж жінки

Вік: Діти та молоді люди мали більші шанси на виживання, ніж люди похилого віку.

Сімейний стан: Пасажири з сім'ями мали більше шансів на виживання ніж самотні пасажири

Місце на кораблі: Пасажири, які знаходились в носовій частині корабля, мали більші шанси на виживання, ніж ті, хто знаходився в кормовій частині.
